```python
import pandas as pd

# Читаем CSV-файл
file_path = "kc_house_data 2 урок.csv"
data = pd.read_csv(file_path)

# Просмотр первых 5 строк
print("Первые 5 строк данных:")
print(data.head())
```

```python
# Информация о данных
print("\nИнформация о данных:")
print(data.info())

# Проверка на пропущенные значения
print("\nКоличество пропущенных значений:")
print(data.isnull().sum())

# Основные статистики
print("\nСтатистическое описание данных:")
print(data.describe())
```

```python
# Типы данных
print("\nТипы данных в каждом столбце:")
print(data.dtypes)

# Анализ диапазонов значений
print("\nДиапазон цен на недвижимость:", data["price"].min(), "-", data["price"].max())
print("Диапазон жилой площади:", data["sqft_living"].min(), "-", data["sqft_living"].max())

# Категориальные признаки
print("\nРаспределение категориальных признаков:")
print("Вид на набережную:\n", data["waterfront"].value_counts())
print("Оценка вида из окон:\n", data["view"].value_counts())
print("Состояние домов:\n", data["condition"].value_counts())
print("Класс дома:\n", data["grade"].value_counts())
```

```python
data.to_csv("processed_kc_house_data.csv", index=False)

# Скачивание файла
files.download("processed_kc_house_data.csv")

```
